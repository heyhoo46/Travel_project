```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
import re
```

```python
def road(home_list, dest_list):
    # Initialize the web driver and navigate to the URL
    google_map_url = f"https://map.naver.com/p/directions/-/-/-/car?c=15.00,0,0,0,dh"
    driver = wb.Chrome()
    driver.get(google_map_url)
    
    # Initialize lists to store the results
    car_time_list = []
    car_distance_list = []
    fare_taxi_list = []
    road_detail_list = []
    
    walk_time_list = []
    walk_distance_list = []
    road_jeju_list = []

    # Wait for elements to load
    searchbox_list = driver.find_elements(By.CSS_SELECTOR, ".input_search")
    button_list = driver.find_elements(By.CSS_SELECTOR, ".btn_direction")

    home_cnt = len(home_list)
    dest_cnt = len(dest_list)

    # Loop through all home and destination pairs
    for i in range(home_cnt):
        home_searchbox = searchbox_list[0]
        home_searchbox.send_keys(Keys.CONTROL, 'A')
        home_searchbox.send_keys(home_list[i])
        time.sleep(1)
        home_searchbox.send_keys(Keys.ENTER)
    
        # Loop through each destination for the current home
        for j in range(dest_cnt):
            dest_searchbox = searchbox_list[1]
            dest_searchbox.send_keys(Keys.CONTROL, 'A')
            dest_searchbox.send_keys(dest_list[j])
            time.sleep(1)
            dest_searchbox.send_keys(Keys.ENTER)
            time.sleep(1)
    
            road_jeju = f"{home_list[i]} -> {dest_list[j]}"
            road_jeju_list.append(road_jeju)
            
            # Click on the "Find Road" button
            button_find_road = button_list[2]
            button_find_road.click()
            time.sleep(2)
    
            # Select the "Car" tab
            btn_car = driver.find_element(By.CSS_SELECTOR, ".sc-szq48w li+li .btn_search_tab")
            btn_car.click()
            time.sleep(2)
            
            ## Car Results ##
            try:
                # Time
                car_time_value = driver.find_element(By.CSS_SELECTOR, ".sc-cdjj58").text
                car_time_list.append(car_time_value)
                
                # Distance
                car_distance_road = driver.find_element(By.CSS_SELECTOR, ".item_distance").text
                car_distance_list.append(car_distance_road)
        
                # Taxi fare
                fare_taxi = driver.find_element(By.CSS_SELECTOR, ".info_fare").text
                fare_taxi_list.append(fare_taxi)
        
                # Road details
                road_detail = driver.find_element(By.CSS_SELECTOR, ".sc-18kdq0f").text
                road_arrow = re.sub(r'(\d+(?:km|m))', r'\1 -> ', road_detail)
                road_arrow = road_arrow.rstrip(" ->")
                road_detail_list.append(road_arrow)
    
            except:
                # In case of errors or no results
                car_time_list.append("")
                car_distance_list.append("")
                fare_taxi_list.append("")
                road_detail_list.append("")
    
            ## Walking Results ##
            btn_walk = driver.find_element(By.CSS_SELECTOR, ".sc-szq48w li+li+li .btn_search_tab")
            btn_walk.click()
            time.sleep(2)
            
            try:
                # Time
                walk_time_value = driver.find_element(By.CSS_SELECTOR, ".sc-cdjj58").text
                walk_time_list.append(walk_time_value)
        
                # Distance
                walk_distance_road = driver.find_element(By.CSS_SELECTOR, ".walk_direction_value").text
                walk_distance_list.append(walk_distance_road)
    
            except:
                # In case of errors or no results
                walk_time_list.append("")
                walk_distance_list.append("")
    
    # Close the driver after the operation
    driver.quit()

    # Return the result lists
    return road_jeju_list, car_time_list, car_distance_list, fare_taxi_list, road_detail_list, walk_time_list, walk_distance_list


```
