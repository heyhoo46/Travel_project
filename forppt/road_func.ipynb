```python
def road(home, dest):

    home_searchbox = searchbox_list[0]
    home_searchbox.send_keys(Keys.CONTROL, 'A')
    home_searchbox.send_keys(home)
    time.sleep(1)
    home_searchbox.send_keys(Keys.ENTER)

    try:
        home_error = driver.find_element(By.CSS_SELECTOR, ".sc-dx2y5q")
        walk_time = "home_error!"
        walk_distance = "home_error!"
        return walk_time, walk_distance
    except:
        
        dest_searchbox = searchbox_list[1]
        dest_searchbox.send_keys(Keys.CONTROL, 'A')
        dest_searchbox.send_keys(dest)
        time.sleep(1)
        dest_searchbox.send_keys(Keys.ENTER)
        time.sleep(1)

        try:
            dest_error = driver.find_element(By.CSS_SELECTOR, ".sc-dx2y5q")
            walk_time = "dest_error!"
            walk_distance = "dest_error!"
            return walk_time, walk_distance
        except:
            # Click on the "Find Road" button
            button_find_road = button_list[2]
            button_find_road.click()
            time.sleep(2)
            
        
            ## Walking Results ##   
            try:
                # Time
                walk_time = driver.find_element(By.CSS_SELECTOR, ".sc-cdjj58").text
                
                # Distance
                walk_distance = driver.find_element(By.CSS_SELECTOR, ".walk_direction_value").text
                
        
            except:
                # In case of errors or no results
                walk_time = "road error!"
                walk_distance = "road error!"
                return walk_time, walk_distance
    # Return the result lists
    return walk_time, walk_distance


```

```python
df_jeju = pd.read_csv("./distance_test.csv")
df_tour = pd.read_csv("./tour_test.csv")
df_hotspot = pd.read_csv("./spot_test.csv")

distance_jeju = df_jeju["distance"]
len(distance_jeju)
home_jeju = df_jeju["home"]
destination_jeju = df_jeju["destination"]
```

```python
home_in1 = []
dest_in1 = []
close_list = []

for i in range(len(distance_jeju)):
    if(distance_jeju[i] < 1):
        close = [home_jeju[i], destination_jeju[i]]
        close_list.append(close)
        home_in1.append(home_jeju[i])
        dest_in1.append(destination_jeju[i])

```

```python
# Initialize the web driver and navigate to the URL
google_map_url = f"https://map.naver.com/p/directions/-/-/-/car?c=15.00,0,0,0,dh"
driver = wb.Chrome()
driver.get(google_map_url)


# Wait for elements to load
searchbox_list = driver.find_elements(By.CSS_SELECTOR, ".input_search")
button_list = driver.find_elements(By.CSS_SELECTOR, ".btn_direction")
btn_walk = driver.find_element(By.CSS_SELECTOR, ".sc-szq48w li+li+li .btn_search_tab")
btn_walk.click()
time.sleep(2)
```

```python
time_walk = []
distance_walk = []
for i in range(len(close_list)):
    road_walk = road(close_list[i][0], close_list[i][1])
    time_walk.append(road_walk[0])
    distance_walk.append(road_walk[1])
```

```python
df_road = pd.DataFrame({"home" : home_in1,
                        "dest" : dest_in1,
                        "time_bywalk" : time_walk,
                        "distance_bywalk" : distance_walk
                       })
df_road.to_csv("testtest.csv", index=False, encoding='utf-8-sig')
```
